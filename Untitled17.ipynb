```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
```

```python
List_Buku = [] #Pada bagian ini, saya membuat sebuah variabel bernama List_Buku yang berbentuk list kosong. List ini nantinya digunakan untuk menampung semua data buku yang berhasil diambil dari website, seperti judul, harga, dan gambar.
```

```python
for page in range(1, 5):  #Di sini saya menggunakan perulangan untuk mengakses beberapa halaman sekaligus, yaitu dari halaman 1 sampai 4. Hal ini dilakukan karena data buku tidak hanya ada di satu halaman saja.
    url = f"https://deepublishstore.com/product-category/ekonomi/page/{page}/" #Kode ini digunakan untuk membuat link website secara otomatis sesuai nomor halaman. Jadi setiap perulangan akan membuka halaman yang berbeda tanpa perlu menulis URL satu per satu.
    response = requests.get(url) #Pada bagian ini, program mengambil isi dari halaman website menggunakan library requests. Hasilnya berupa kode HTML dari halaman tersebut.
    soup = BeautifulSoup(response.text, "html.parser")

    judul = soup.find_all("h2", class_="woocommerce-loop-product__title") #Di sini saya mengambil semua judul buku yang ada di halaman. Tag yang digunakan adalah <h2> dengan class tertentu sesuai struktur HTML website.
    harga = soup.find_all("span", class_="woocommerce-Price-amount amount") #Kode ini digunakan untuk mengambil harga dari setiap buku. Data harga biasanya berada dalam tag <span> dengan class tertentu.
    gambar = soup.find_all("img", class_="attachment-woocommerce_thumbnail size-woocommerce_thumbnail") #Bagian ini berfungsi untuk mengambil link gambar dari setiap buku. Data gambar diambil dari tag <img>.

    for i in range(len(judul)): #Setelah semua data diambil, dilakukan perulangan untuk menggabungkan data judul, harga, dan gambar menjadi satu kesatuan per buku.
        data = {
            "Judul": judul[i].text.strip(),
            "Harga": harga[i].text.strip() if i < len(harga) else None,
            "Gambar": gambar[i]['src'] if i < len(gambar) else None
        }
        List_Buku.append(data)
```

```python
with open("List_Buku.json", "w", encoding="utf-8") as file: #Setelah semua data terkumpul, data disimpan ke dalam file JSON. Format JSON dipilih karena mudah dibaca dan sering digunakan untuk pertukaran data.
    json.dump(List_Buku, file, ensure_ascii=False, indent=4)
```

```python
df = pd.DataFrame(List_Buku) #Di sini data yang sudah berbentuk list diubah menjadi DataFrame agar lebih mudah diolah dan disimpan ke format lain seperti CSV.
df.to_csv("List_Buku.csv", index=False) #Terakhir, data disimpan ke dalam file CSV. Format ini biasanya digunakan untuk keperluan analisis data karena bisa dibuka di Excel.
```
